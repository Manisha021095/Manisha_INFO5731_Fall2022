# **Question 2**

(30 points). Write a python program to **clean the text data** you collected above and save the data in a new column in the csv file. The data cleaning steps include:

(1) Remove noise, such as special characters and punctuations.

(2) Remove numbers.

(3) Remove stopwords by using the [stopwords list](https://gist.github.com/sebleier/554280).

(4) Lowercase all texts

(5) Stemming. 

(6) Lemmatization.

```python
# Write your code here

import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
# 1,2,4  

def text_clean(text):
    clean_text = text.str.lower().str.replace('[^a-z ]', '')
    return clean_text
```

```python
# 3

stop_words = set(stopwords.words('english'))

def stopword(text):
    new_text = " ".join([x for x in text.split() if not x in list(map(str.lower,stop_words))])
    return new_text
```

```python
#5

stemmer = PorterStemmer()

def PS_stemmer(text):
    new_text = " ".join([stemmer.stem(x) for x in text.split()])
    return new_text
```

```python
#6

lemmatizer = WordNetLemmatizer()

def WN_lemmatizer(text):
    new_text = " ".join([lemmatizer.lemmatize(x, pos = 'a') for x in text.split()])
    return new_text
```

```python
# Application of above defined functions

# "text_clean"

mod_text = text_clean(Data["Reviews"])
```

```python
# "stopword"

mod_text = mod_text.apply(stopword)
```

```python
# "PS_stemmer"

mod_text = mod_text.apply(PS_stemmer)
```

```python
# "WN_lemmatizer"

mod_text = mod_text.apply(WN_lemmatizer)
```

```python
Data["Clean Reviews"] = mod_text
```
